## Crossreferenceable “Table”s

Tables with a caption and a label can be referenced using the `@ref`
syntax, as <a href="#tbl-1" class="quarto-xref">Table 1</a> shows.

| Default | Left | Right | Center |
|---------|:-----|------:|:------:|
| 12      | 12   |    12 |   12   |
| 123     | 123  |   123 |  123   |
| 1       | 1    |     1 |   1    |

Table 1: This is the caption for the table

Tables can contain arbitrary content instead of an image. This lets you,
for example, use an image to represent the table, in case its formatting
is too complex for it to be rendered in HTML.
